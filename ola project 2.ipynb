# Univariate Analysis

Employee Demographics

```python
Hr_data.columns
```

```python
plt.figure(figsize=(10,6))
ax=sns.countplot(x='MaritalStatusID', data=Hr_data1)
plt.title('Marital Status Distribution')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.xticks([0, 1, 2, 3, 4], ['Single','Married','Divorced','Seperated','Widowed'])
for p in ax.patches:
    ax.annotate(f'{p.get_height()}',(p.get_x()+p.get_width()/2.,p.get_height()),
ha='center',va='center',xytext=(0,5),textcoords='offset points')
plt.show()
```

#Inference:The Marital status distribution shows tha majority of employees in the organisation are single and married, while the count of widowed employees is the lowest. It is possible that the organisation attracts a larger number of single and married people due to factors like nature of the industry, work-life balance policies or geographical location.

```python
Gender_count=Hr_data1["GenderID"].value_counts()
plt.figure(figsize=(8, 5))
plt.pie(Gender_count, labels=Gender_count.index, autopct='%1.1f%%', startangle=0)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Add a title
plt.title("Gender Distribution")
labels= ['Female', 'Male']
plt.legend(labels)
# Show the plot
plt.show()
```

#Inference:The organisation has a higher representation of female employees, accounting for 56.6% of the total workforce. This gender distribution could be influenced by various factors such as recruitment strategies, workplace diversity, initiatives or industry trends.

Employee Analysis

```python
Dept_count=Hr_data1["Department"].value_counts()
plt.figure(figsize=(8, 5))
plt.pie(Dept_count, labels=Dept_count.index, autopct='%1.1f%%', startangle=0)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Add a title
plt.title('Employee Distribution Across Department')

# Show the plot
plt.show()
```

#Inference:The production department has the largest population of the workforce in the organisation, while the Admin Office and Executive office have a relatively smaller workforce compared to other departments in the industry. This distribution could be due to the nature of the organisation's operations, where the production department requires a larger workforce to handle manufacturing or operational tasks,while the admin and executive offices may have a smaller team to manage administrative and managerial responsibilities.

```python
sns.countplot(y='RecruitmentSource',data=Hr_data1)
plt.title('Recruitment Source Distribution')
plt.ylabel('Recruitment Sources')
plt.xlabel('Count')
```

#Inference:The organisation primarily uses Indeed as the most popular source for sourcing employees, followed by linkedin.It is interesting to see that online web applications and other sources of employment are rarely used.Leveraging popular job serch platforms like Indeed and linkedin can be effective in reaching a wide pool of potential candidated. However, it is important to consider other recruitment startegies as well to ensure a diverse and well rounded applicant pool.

```python
fig=plt.figure(figsize=(8,8))
sns.countplot(y='TermReason',hue='Department',data=Hr_data1)
plt.title('Reason for Termination')
plt.ylabel('Termination Reasons')
plt.xlabel('Count')
```
